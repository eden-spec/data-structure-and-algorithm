    *   Implementing a Bubble Sort algorithm to sort the list of student dictionaries by ID.
    *   Implementing an Insertion Sort algorithm to sort the list of student dictionaries alphabetically by Name.
    *   Implementing a Selection Sort algorithm to sort the list of student dictionaries by CGPA in descending order.

```python

students = [
    {'id': 'DBU1700003', 'name': 'Estifanos', 'dept': 'DS', 'cgpa': 3.74},
    {'id': 'DBU1700007', 'name': 'Abrham', 'dept': 'CS', 'cgpa': 3.52},
    {'id': 'DBU1700001', 'name': 'Biruk', 'dept': 'IS', 'cgpa': 2.94},
    {'id': 'DBU1700009', 'name': 'Rahel', 'dept': 'IS', 'cgpa': 3.10},
    {'id': 'DBU1700005', 'name': 'Mahlet', 'dept': 'IT', 'cgpa': 3.45},
    {'id': 'DBU1700002', 'name': 'Selam', 'dept': 'IT', 'cgpa': 2.59},
    {'id': 'DBU1700006', 'name': 'Yonatan', 'dept': 'SE', 'cgpa': 3.67},
    {'id': 'DBU1700010', 'name': 'Helen', 'dept': 'CS', 'cgpa': 3.80},
    {'id': 'DBU1700008', 'name': 'Kibrom', 'dept': 'DS', 'cgpa': 2.25},
    {'id': 'DBU1700004', 'name': 'Meron', 'dept': 'SE', 'cgpa': 3.06}
]
```

### Bubble Sort: Sorts the students by 'id' in ascending order

```python
def bubble_sort_by_id(student_list):
    n = len(student_list)
    for i in range(n):
        for j in range(0, n - i - 1):
            if student_list[j]['id'] > student_list[j + 1]['id']:
                student_list[j], student_list[j + 1] = student_list[j + 1], student_list[j]
    return student_list

sorted_by_id = bubble_sort_by_id(students.copy())
print("Students sorted by ID (Bubble Sort):")
for student in sorted_by_id:
    print(student)
```

### Insertion Sort: Sorts the students alphabetically by 'name'

```python

def insertion_sort_by_name(student_list):
    for i in range(1, len(student_list)):
        key = student_list[i]
        j = i - 1
        while j >= 0 and key['name'].lower() < student_list[j]['name'].lower():
            student_list[j + 1] = student_list[j]
            j -= 1
        student_list[j + 1] = key
    return student_list

sorted_by_id = insertion_sort_by_name(students.copy())
print("Students sorted by Name (Insertion Sort):")
for student in sorted_by_id:
    print(student)
```

### Selection Sort: Sorts the students by 'cgpa' in descending order



```python
def selection_sort_by_cgpa(student_list):
    n = len(student_list)
    for i in range(n):
        max_idx = i
        for j in range(i + 1, n):
            if student_list[j]['cgpa'] > student_list[max_idx]['cgpa']:
                max_idx = j
        student_list[i], student_list[max_idx] = student_list[max_idx], student_list[i]
    return student_list

sorted_by_id = selection_sort_by_cgpa(students.copy())
print("Students sorted by CGPA (insertion Sort):")
for student in sorted_by_id:
    print(student)
```

### <p style="font-size: 14pt">Analysis</p>

#### <p style="font-size: 12pt">Performance Comparison</p>

The code showcases three sorting algorithms: `bubble_sort_by_id`, `insertion_sort_by_name`, and `selection_sort_by_cgpa`. While all three have a worst-case and average-case time complexity of O(n^2), their actual performance will vary depending on the initial state of the student data list.

*   Bubble Sort: The `bubble_sort_by_id` function is designed to arrange student dictionaries in ascending order based on the 'id' field. It iterates multiple times to compare each adjacent student's id in the list, swapping elements that are out of order. It is expected to perform more comparisons and swaps, resulting in its low performance.

*   Insertion Sort: The `insertion_sort_by_name` function sorts student dictionaries alphabetically by 'name'. For each element, it creates a sorted sub-list to insert elements by the value of name. This involves fewer comparisons than bubble sort however shifting operations still result in O(n^2) on average.

*   Selection Sort: The `selection_sort_by_cgpa` arranges student dictionaries in descending order based on their 'cgpa' values. Selection Sort tends to minimize swap operations, providing it with a relatively consistent performance, regardless of the input.

#### <p style="font-size: 12pt">Advantages and Disadvantages for Sorting Dictionary Data</p>

Here are the advantages and disadvantages for sorting premade Python dictionary datasets.

*   Bubble Sort:
    *   Advantages: Simple to implement and in-place sorting (requiring minimal additional memory).
    *   Disadvantages: Inefficient for large datasets due to quadratic time complexity, resulting in slower performance compared to insertion sort and selection sort.

*   Insertion Sort:
    *   Advantages: Relatively simple implementation, efficiency on small or nearly sorted datasets, and in-place sorting capabilities.
    *   Disadvantages: Due to the high number of loops required, it still has a high average complexity time.

*   Selection Sort:
    *   Advantages: Simple to implement and performs fewer swaps compared to Bubble Sort, improving its efficiency for larger datasets.
    *   Disadvantages: Limited by its O(n^2) time complexity.

*Overall summary of the three algorithms, selection sort algorithm is the most efficient because of the low number of swaps. Insertion sort and bubble sort both have relatively high complexity and loops resulting in a lower performance algorithm.*
